# 35 Numeric Pattern Programs in Python

```python
# Pattern - 1
for i in range(5):
    for j in range(i+1):
        print(i+1, end=' ')
    print("\n")
        
```

```python
# Pattern - 32
for i in range(5):
    p=i
    for j in range(5):
        if (j+1)%2!=0:
            p+=1
            print(p,end=' ')
        else:
            p+=9
            print(p-2*i,end=' ')
    print('\n')
        
```
